Var
   // Seção de Declarações das variáveis
   GNV,COMB,PNEU,OLEO,DESC_COMB,IPVA,KITG,LIMP,RACK,VL_COMB,VL_TX,TOTAL,VL_FRETE,LUCRO:REAL
   KM,DOBRA :INTEIRO

Inicio
   // Seção de Comandos, procedimento, funções, operadores, etc...

   ESCREVAL ("DIGITE A DISTÂNCIA ATÉ O LOCAL")
   LEIA (KM)
   DOBRA <- KM*2  //VALOR DOBRA SE TRATANDO DE IDA E VOLTA
   ESCREVAL ("                   >>   ATENÇÃO !!!  <<  ")
   ESCREVAL (" - O KM INFORMADO REPRESENTARÁ"," IDA E VOLTA"," AO LOCAL DE DESTINO -")
   ESCREVAL (" - O CAÚCULO FOI  BASEADO NO VALOR DE  ", DOBRA ," QUILOMETROS - ")
   ESCREVAL("=================================================================")
   ESCREVAL ("TABELA DE DESPESAS")
   ESCREVAL("=================================================================")
   GNV <- (0.40 * DOBRA) //VALOR BASEADO EM 60.00 REAIS REFERENTE AO CILINDRO CHEIO,
   //PODENDO ATINGIR 160 QUILOMETROS COM OS 60 REAIS
   ESCREVAL ("  #  VALOR A SER GASTO EM GÁS NATURAL",GNV)
   COMB <- (0.60 * DOBRA) //REFERENTE A DIVISÃO DE 12 KM/LITRO SENDO O COMBUSTIVEL
   //7 REAIS (PREÇO DA GASOLINA NA ATUALIDADE )
   DESC_COMB <- (0.3 * COMB)  // SABENDO QUE O GNV TRABALHA SECO DEVE SER UTILIZADO
   //A GASOLINA PARA LUBRIFICAR TODO O SISTEMA (VALOR DE 30% REFERENTE AO GNV USADO
   //AUXILIANDO EM TRECHOS DE SUBIDAS INGRIMES
   ESCREVAL ("  #  VALOR A SER GASTO EM COMBUSTÍVEL",DESC_COMB)
   VL_COMB <- GNV + DESC_COMB   //SE REFERE A REGRA UTILIZADA
   ESCREVAL("                                 >>>>",VL_COMB," REAIS"," <<<<")
   ESCREVAL("=================================================================")
   PNEU <- (0.10 * KM )    //BASEADO NO CUSTO TOTAL DE 2500 REAIS PARA TROCA
   //ALINHAMENTO E BALANCEAMENTO , RETIRAR E COLOCAR NO LUGAR (2500/50000KMS RODADOS)
   ESCREVAL ("  #  TX DESTINADA A MANUTENÇÃO >> PNEUS <<",PNEU," REAIS,.")
   OLEO <- (0.10) * KM  //TAXA BASEADA EM 200 REAIS A CADA 10000
   ESCREVAL ("  #  TX DESTINADA A MANUTENÇÃO >> OLÉO <<",OLEO," REAIS,.")
   LIMP <- 20.00  // O FRENTE SEMPRE DEIXA RESIDUOS DE DOCES , NADA MELHOR QUE
   //LAVAGEM DO CARRO POR FRETE REALIZADO.
   ESCREVAL ("  #  TX DESTINADA A LIMPEZA",LIMP," REAIS.")
   KITG <- 15.00  //O VALOR FOI DESTINADO A AQUISIÇÃO DO GNV 5000/365 DIAS .
   ESCREVAL ("  #  TX DESTINADA AO KIT GÁS",KITG," REAIS,.")
   RACK <- 4.00 //VALOR FOI DESTINADO A AQUISIÇÃO DA INSTALAÇÃO DO RACK
   //800 REAIS DIVIDIDO POR 365 DIAS
   ESCREVAL ("  #  TX DEDUZIDA  AO RACK",RACK," REAIS,.")
   IPVA <- 6.00 //VALOR FOI DESTINADO AO IPVA DO CARRO
   //1500/365 DIAS
   ESCREVAL ("  #  VALOR DEDUZIDO AO IPVA",IPVA," REAIS, .")
   ESCREVAL("=================================================================")
   VL_TX <- ( PNEU + OLEO + KITG + RACK +IPVA ) // SOMA DAS DESPESAS
   ESCREVAL ("   VALOR DESTINADO A MANUTENÇÃO >>>>",  VL_TX ," <<<<"," REAIS.")
   ESCREVAL("================================================================")
   TOTAL <-(VL_COMB + VL_TX)
   ESCREVAL ("                        TOTAL GERAL DE DESPESAS",TOTAL, " REAIS.")
   ESCREVAL("=================================================================")
   VL_FRETE <- TOTAL * 2
   SE VL_FRETE <= 149.99 ENTAO
      ESCREVAL ("             VALOR A SER COBRADO NO FRETE ","150.00"," REAIS.")
      ESCREVAL("====================================================================")
      LUCRO <- (150.00 - TOTAL )
      ESCREVAL ("                O LUCRO PARA O FRETE É DE ",LUCRO  , " REAIS.")
      ESCREVAL("===================================================================")
   SENAO
      VL_FRETE <- TOTAL * 2
      ESCREVAL ("               VALOR A SER COBRADO NO FRETE ",VL_FRETE," REAIS.")
      ESCREVAL("==============================================================")
      LUCRO <- (VL_FRETE - TOTAL )
      ESCREVAL ("                  O LUCRO PARA O FRETE É DE ",LUCRO  , " REAIS.")
      ESCREVAL("=====================================================================")
   FIMSE

   ESCREVAL (".")
   ESCREVAL (".")
